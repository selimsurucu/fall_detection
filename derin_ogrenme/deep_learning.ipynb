# Dataları Birleştirme

```python
data=pd.concat([data_bsc,data_fkl,data_fol,data_sch,data_sdl,data_std,data_stn,data_stu,data_wal])
data['type'] = data['type'].astype('category')
data['type'] = data['type'].cat.codes

data.tail(100)
```

```python
data.index = data.time
```

```python
data
```

```python
data.drop('time', axis=1, inplace=True)
```

```python
data['x(t+1)'],data['y(t+1)'],data['z(t+1)'] = data.x.shift(1),data.y.shift(1),data.z.shift(1)
data.head()
```

```python
data['x(t+2)'],data['y(t+2)'],data['z(t+2)'] = data.x.shift(2),data.y.shift(2),data.z.shift(2)
data.head()
```

```python
data = data.dropna()
```

```python
data
```

```python
print(data.type)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    data.drop('type', axis=1), data.type, test_size=0.33, random_state=42)
```

```python
from tensorflow.keras import models, layers
import sklearn.metrics as metrics
```

```python
nn2 = models.Sequential()
nn2.add(layers.Dense(128, input_shape=(X_train.shape[1],), activation='relu'))
nn2.add(layers.Dense(256, activation='relu'))
nn2.add(layers.Dense(256, activation='relu'))
nn2.add(layers.Dense(256, activation='relu'))
nn2.add(layers.Dense(256, activation='relu'))
nn2.add(layers.Dense(1, activation='linear'))

# Compiling the model
nn2.compile(loss='mean_squared_error',
            optimizer='adam',
            metrics=['mean_squared_error'])

# Printing the model summary
print(nn2.summary())
```

```python
n2_history = nn2.fit(X_train,
                  y_train,
                  epochs=100,
                  batch_size=256,
                  validation_split = 0.1)
```

```python
y_test_pred = nn2.predict(X_test)
y_train_pred = nn2.predict(X_train)
print("Training R squared:", round(metrics.r2_score(y_train, y_train_pred),4))
print("Validation R squared:", round(metrics.r2_score(y_test, y_test_pred),4))
```
